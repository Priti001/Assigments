```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

#Create dataframes with files
school= pd.read_csv('C:\\Pri doc\\priti\\Projects\\Myproject\\Berkely\\Week 3 Python\\schools_complete.csv')
student= pd.read_csv('C:\\Pri doc\\priti\\Projects\\Myproject\\Berkely\\Week 3 Python\\students_complete.csv')
```

```python
#Create a high level snapshot (in table form) of the district's key metrics, including:
#Total Schools
Total_schools=len(school.index)
#Total Students
Total_stu=len(student.index)
#Total Budget
Total_bud= school['budget'].sum()
#Average Math Score
Avg_math=student['math_score'].mean()
round(Avg_math,2)
#Average Reading Score
Avg_readscore=round(student['reading_score'].mean(),2)
```

```python
#% Passing Math (passing requires a score of 70% or greater)
Per_Passing_math=(student['math_score'] > 70 ).value_counts()[True]/Total_stu*100
#% Passing Reading (passing requires a score of 70% or greater)
Per_Passing_reading=(student['reading_score'] > 70 ).value_counts()[True]/Total_stu*100
#Overall Passing Rate (Average of the above two)
Overall_passing_rate= (Per_Passing_math+Per_Passing_reading)/2
Per_Passing_math
```

```python
#Create dictionary and set to dataframe
district_sum = {'Total Schools': Total_schools,
                'Total Students': Total_stu,
               'Total Budget':Total_bud,
                'Average Math Score':Avg_math,
                'Average Reading Score':Avg_readscore,
                '% Passing Math':Per_Passing_math,
                '% Passing Reading':Per_Passing_reading,
                'Overall Passing Rate':Overall_passing_rate
               }
district_summary = pd.DataFrame([district_sum], columns=district_sum.keys())
district_summary 
```

```python
#Student count for Math_score>70 per school
x = student[student['math_score'] > 70]
x['school'].value_counts()
```

```python
#Create new dataframe with groupby on school with student data
df = student.groupby(['school']).agg({'Student ID':['count'],'reading_score':['mean'],'math_score':['mean']})
```

```python
df.columns = df.columns.droplevel(1)
df.columns
```

```python
df['%Passing_math']=student[student['math_score'] > 70].school.value_counts()/student['school'].value_counts()*100
```

```python
df['%Passing_reading_score']=student[student['reading_score'] > 70].school.value_counts()/student['school'].value_counts()*100
```

```python
df['Overall_passing_rate']= df[['%Passing_math', '%Passing_reading_score']].mean(axis=1)
```

```python
df
```

```python
df=df.reset_index()
```

```python
school.rename(columns={'name': 'school'},inplace=True)
```

```python
df.rename(columns={'Student ID': 'total_student'},inplace=True)
df
```

```python
#merge the data to get school and student together
merged_data = pd.merge(df,school,on='school')
```

```python
#Add new column to get per_student_budget
merged_data['Per_student_budget']= merged_data['budget']/merged_data['total_student']
merged_data['Per_student_budget'] = merged_data['Per_student_budget'].astype(int)
merged_data
```

```python
#Top performing schhol by passing rate
merged_data.nlargest(5, 'Overall_passing_rate')
```

```python
#Bottom performing schools by passing rate
merged_data.nsmallest(5, 'Overall_passing_rate')
```
